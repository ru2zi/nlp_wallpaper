# 모델지정

```python
# langchain_community 라이브러리에서 HuggingFaceEmbeddings를 가져옵니다.
from langchain_community.embeddings import HuggingFaceEmbeddings

# HuggingFace의 "distiluse-base-multilingual-cased-v1" 모델 경로를 지정합니다.
modelPath = "distiluse-base-multilingual-cased-v1"

# 모델 및 인코딩 관련 매개변수를 설정합니다.
model_kwargs = {'device': 'cuda'}  # GPU 가속을 사용하도록 설정합니다.

# 인코딩 시 사용할 매개변수를 설정합니다. 여기서는 embeddings를 정규화하지 않도록 설정했습니다.
encode_kwargs = {'normalize_embeddings': False}

# HuggingFaceEmbeddings 객체를 생성합니다.
embeddings = HuggingFaceEmbeddings(
    model_name=modelPath,       # 사용할 HuggingFace 모델의 이름 또는 경로를 지정합니다.
    model_kwargs=model_kwargs,  # 모델 초기화에 사용될 매개변수를 전달합니다.
    encode_kwargs=encode_kwargs  # 인코딩에 사용될 매개변수를 전달합니다.
)
```

# 모델 id지정 후 파이프라인 생성

```python
# langchain_community 라이브러리에서 HuggingFacePipeline을 가져옵니다.
from langchain_community.llms.huggingface_pipeline import HuggingFacePipeline

# transformers 라이브러리에서 GPT2LMHeadModel, AutoTokenizer, pipeline을 가져옵니다.
from transformers import GPT2LMHeadModel, AutoTokenizer, pipeline

# 사용할 Hugging Face 모델의 ID를 지정합니다.
model_id = "EleutherAI/polyglot-ko-1.3b"

# AutoTokenizer를 사용하여 토크나이저를 불러오고, GPT2LMHeadModel을 사용하여 모델을 불러옵니다.
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = GPT2LMHeadModel.from_pretrained(model_id).to(0)  # 모델을 GPU(장치 0)에 로드합니다.

# pipeline 함수를 사용하여 텍스트 생성을 위한 파이프라인을 생성합니다.
# max_new_tokens는 생성할 최대 토큰 수를 나타냅니다. device는 사용할 GPU 장치를 나타냅니다.
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=512, device=0)

# HuggingFacePipeline 객체를 생성합니다.
hf = HuggingFacePipeline(pipeline=pipe)
```

# 답변출력

```python
# langchain_core 라이브러리에서 PromptTemplate, RunnablePassthrough, StrOutputParser를 가져옵니다.
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

# format_docs 함수는 주어진 문서들을 페이지 내용으로 포맷하여 반환합니다.
def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs)

# RAG(Retrieval-Augmented Generation) 모델의 템플릿으로 사용할 문자열 템플릿을 정의합니다.
template = """마지막에 질문에 답하려면 다음과 같은 맥락을 사용합니다.

{context}

질문: {question}

유용한 답변:"""

# PromptTemplate 객체를 생성하고, 문자열 템플릿을 템플릿으로 사용합니다.
custom_rag_prompt = PromptTemplate.from_template(template)

# RAG 체인을 설정합니다.
rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}  # 문맥을 가져오는 retriever를 사용하고, format_docs 함수로 문서를 포맷합니다.
    | custom_rag_prompt  # 사용자 정의 템플릿을 적용합니다.
    | hf  # HuggingFacePipeline을 사용하여 텍스트를 생성합니다.
    | StrOutputParser()  # 생성된 텍스트 결과를 문자열로 파싱합니다.
)
```

```python
for chunk in rag_chain.stream("도배지에 녹은 자국이 발생하는 주된 원인과 그 해결 방법은 무엇인가요?"):
    print(chunk, end="", flush=True)
```

# 정답레이블 비교 후 피클 저장

```python
test = pd.read_csv("/content/drive/MyDrive/Dacon_test/test.csv")
```

```python
result = []

# 2048 길이 제한에 걸림 길이 제한을 줘야할듯함 일단 성능 넘 안나와서
# 이거는 살려두고 다른코랩에서 작성해봄

for i in tqdm(range(len(test))):
  _id = test.at[i,'id']
  _q = test.at[i,'질문']
  _a = []
  for chunk in rag_chain.stream(_q):
      _a.append(chunk)
      print(chunk, end="", flush=True)
  result.append(
      {
          "id":_id,
          "대답":" ".join(_a)
      }
  )
  print()
```
